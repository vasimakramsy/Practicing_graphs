```
'''importing dependencies'''

import matplotlib.pyplot as plt
import numpy as np
```

```
np.random.seed(30)

X = np.random.randint(1,100,100)

y = 3 + 4*X
```

```
# ploting a scatter graph

plt.scatter(X,y)
```

```
# lets plot bar graph

plt.bar(X,y)

plt.xlabel('x-value')

plt.ylabel('y-value')
```

```
# lets plot box plot usally we use box plot to see the where the data-points lies more in dataset

plt.boxplot(y)
```

```
'''creating a ladder'''

plt.plot([1,2,3],[3,8,18],linestyle=':')
plt.show()
```

```
'''creating mark at begining,middle and end'''

plt.plot([1,2,3],[3,8,18],marker='+')

plt.show()
```

```
'''using figsize we can change the size of graph and changed the points'''
plt.figure(figsize=(5,10))
plt.plot([1,2,3],[2,4,9],marker='.',markerfacecolor='red',markersize=13.0)
plt.show()
```

```
'''plotting 2 lines at a time here linewidth will define the size of line'''
plt.plot([1,2,3],[3,8,18],marker='+',linewidth=4)
plt.plot([1,2,8],[4,7,11],linewidth=4)
plt.show()
```

```
'''lets add grids to same above graph (grids provide more deatils)'''
plt.grid()
plt.plot([1,2,3],[3,8,18],marker='+',linewidth=4)
plt.plot([1,2,8],[4,7,11],linewidth=4)
plt.show()
```

```
'''lets see same graph in linesteps'''
plt.grid()
plt.plot([1,2,3],[3,8,18],marker='+',linewidth=4,linestyle='steps')
plt.plot([1,2,8],[4,7,11],linewidth=4,linestyle='steps')
plt.show()
```

```
'''plotting with different shape'''
# plt.figure(figsize=(10,10))
p=np.arange(0,5,0.2)
plt.plot(p,p,'r--',p,p**2,'b^',p,p**3,'gs')
plt.show()
```

```
'''plotting for catagorical data
plotting subplots and all graphs in same line
note:subplots should have 3 digits
'''
names=["Apple","Mango","Grape","Cherry"]
values=[10,200,300,500]
plt.figure(1,figsize=(9,3))

plt.subplot(131)
plt.bar(names,values)

plt.subplot(132)
plt.scatter(names,values)

plt.subplot(133)
plt.plot(names,values)

plt.show()
```
